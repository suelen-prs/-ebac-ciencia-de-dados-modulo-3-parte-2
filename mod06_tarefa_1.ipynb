```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig, axes = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.4, hspace=0.2)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=sisnac)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=sisnac)
sns.countplot(ax = axes[1, 0], x='SEXO', data=sisnac)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=sisnac)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
X = list(sisnac['APGAR5'])
Y = list(sisnac['CONSULTAS'])

plt.bar(X, Y, color='g')
plt.title("APGAR5 X CONSULTAS")
plt.xlabel("APGAR5")
plt.ylabel("CONSULTAS")

plt.show()
```
