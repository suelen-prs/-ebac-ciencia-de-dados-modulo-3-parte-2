#### 1. Carregue os dados da tarefa anterior

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
dados = pd.read_csv('/content/yahoo_finance_ativos_ex01.csv')
dados.head()
````

````python
# Limpeza dos dados
dados_limpados = dados.drop([0, 1]).reset_index(drop=True)
dados_limpados.columns = [
    'Date', 'Adj Close CSCO', 'Adj Close IBM', 'Adj Close ORCL',
    'Close CSCO', 'Close IBM', 'Close ORCL',
    'High CSCO', 'High IBM', 'High ORCL',
    'Low CSCO', 'Low IBM', 'Low ORCL',
    'Open CSCO', 'Open IBM', 'Open ORCL',
    'Volume CSCO', 'Volume IBM', 'Volume ORCL'
]
print("Dados após limpeza e ajuste de cabeçalhos:")
dados_limpados.head()
````

````python
# Conversão dos tipos de dados
dados_limpados['Date'] = pd.to_datetime(dados_limpados['Date'])
for col in ['Adj Close CSCO', 'Adj Close IBM', 'Adj Close ORCL']:
    dados_limpados[col] = pd.to_numeric(dados_limpados[col])
print("\nTipos de dados após conversão:")
dados_limpados.dtypes.head()
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# Cálculo da média móvel de 90 dias
for ticker in ['CSCO', 'IBM', 'ORCL']:
    dados_limpados[f'MA90 {ticker}'] = dados_limpados[f'Adj Close {ticker}'].rolling(window=90).mean()

# Exibir os resultados da média móvel de 90 dias
print("\nMédia Móvel de 90 dias para os ativos:")
dados_limpados[['Date', 'MA90 CSCO', 'MA90 IBM', 'MA90 ORCL']].tail()
````

````python
# Cálculo do desvio padrão móvel de 90 dias
for ticker in ['CSCO', 'IBM', 'ORCL']:
    dados_limpados[f'STD90 {ticker}'] = dados_limpados[f'Adj Close {ticker}'].rolling(window=90).std()

# Exibir os resultados do desvio padrão móvel de 90 dias
print("\nDesvio Padrão Móvel de 90 dias para os ativos:")
dados_limpados[['Date', 'STD90 CSCO', 'STD90 IBM', 'STD90 ORCL']].tail()
````

````python
# Geração dos gráficos
# Gráfico das médias móveis
plt.figure(figsize=(14, 5))
for ticker in ['CSCO', 'IBM', 'ORCL']:
    plt.plot(dados_limpados['Date'], dados_limpados[f'MA90 {ticker}'], label=f'{ticker} MA90')
plt.title('Média Móvel de 90 Dias dos Ativos')
plt.xlabel('Data')
plt.ylabel('Preço Médio')
plt.legend()
plt.show()
````

````python
# Gráfico dos desvios padrão
plt.figure(figsize=(14, 5))
for ticker in ['CSCO', 'IBM', 'ORCL']:
    plt.plot(dados_limpados['Date'], dados_limpados[f'STD90 {ticker}'], label=f'{ticker} STD90')
plt.title('Desvio Padrão Móvel de 90 Dias dos Ativos')
plt.xlabel('Data')
plt.ylabel('Desvio Padrão')
plt.legend()
plt.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Preparação dos dados
dados_limpados['Date'] = pd.to_datetime(dados_limpados['Date'])
ativos_cols = ['Open CSCO', 'High CSCO', 'Low CSCO', 'Close CSCO']
dados_candle = dados_limpados[-60:][['Date'] + ativos_cols].copy()
dados_candle['Date'] = mdates.date2num(dados_candle['Date'])  # Converte datas para o formato numérico

# Assegurando que os dados de preço são numéricos
for col in ['Open CSCO', 'High CSCO', 'Low CSCO', 'Close CSCO']:
    dados_candle[col] = pd.to_numeric(dados_candle[col])

fig, ax = plt.subplots(figsize=(15, 7))

# Função para desenhar o gráfico de candlestick
def plot_candlestick(df, ax):
    for i, row in df.iterrows():
        color = 'green' if row['Close CSCO'] >= row['Open CSCO'] else 'red'
        # Linha do candle
        ax.plot([row['Date'], row['Date']], [row['Low CSCO'], row['High CSCO']], color=color)
        # Corpo do candle
        inicio = row['Date'] - 0.1
        altura = abs(row['Close CSCO'] - row['Open CSCO'])
        ax.add_patch(plt.Rectangle((inicio, min(row['Open CSCO'], row['Close CSCO'])),
                                   0.2, altura, color=color, alpha=0.5))

plot_candlestick(dados_candle, ax)

# Formatando o gráfico
ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)
plt.title("Gráfico de Candlestick para CSCO - Últimos 60 dias")
plt.xlabel("Data")
plt.ylabel("Preço")
plt.grid(True)

plt.show()
````
