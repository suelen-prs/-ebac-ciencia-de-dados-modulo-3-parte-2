#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
import pandas as pd
import numpy as np
import yfinance as yf
import plotly.express as px


from pandas_datareader import data as pdr
from datetime import date
from plotly import graph_objects
```

```python
from datetime import date, timedelta

# Pegar a data de hoje
hoje = date.today()

# Pegar a data do dia anterior
dia_anterior = hoje - timedelta(days=1)

print("Data de hoje:", hoje)
print("Data do dia anterior:", dia_anterior)
```

```python
# lista de ativos (Intel, Dell e Acer)
ativos = ['CSCO', 'ORCL', 'IBM']

# pega a data atual - 2 anos
data_ini = str(dia_anterior.replace(year = date.today().year - 2))

# baixo os dados do Yahoo Finance
yf.pdr_override()
df = yf.download(ativos, start=data_ini)

# salvo os dados
df.to_csv('yahoo_finance_ativos_ex01.csv')
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# Dataframe com valores minimo e máximo
df_min = df['Low'].unstack().reset_index()
df_max = df['High'].unstack().reset_index()

# Novo dataframe a partir do dataframe de valores minimos
df_ativos = df_min.copy()

# renomeio as colunas
df_ativos.columns = ['ativo', 'data', 'valor_minimo']

# adiciono a coluna com os valores maximos
df_ativos['valor_maximo'] = df_max[0]

# calculo a variação
df_ativos['variacao_valor'] = df_ativos['valor_maximo'] - df_ativos['valor_minimo']

df_ativos
```
